```python
import matplotlib.pyplot as plt
import numpy as np

# Load data from the file
data = np.loadtxt('dos-la2o3+u.txt')  # Assumes a space-separated text file

# Extract Energy and TDOS columns
energy = data[:, 0] 
tdos = data[:, 1]

# Create a plot
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed

# Plot the TDOS data
plt.plot(energy, tdos, label='TDOS', color='navy')

# Add a vertical dotted line for Fermi Energy
fermi_energy = 0.0  # Change this value to the actual Fermi Energy
plt.axvline(x=fermi_energy, color='gray', linestyle='--', label='Fermi Energy')

plt.xlabel('E - Ef (eV)', fontsize=16)
plt.ylabel('DOS (states/eV)', fontsize=16)
plt.title('Density Of States La2O3 (DFT+U)', fontsize=20)

# Add a dotted grid
plt.grid(True, linestyle='--')

plt.legend()

plt.show()  # Show the plot
```
