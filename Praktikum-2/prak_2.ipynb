```python
import matplotlib.pyplot as plt
import numpy as np

# Load data from the file
data = np.loadtxt('dos_C.txt')  # Assumes a space-separated text file
title = "Density of States (DOS) C"

# Extract Energy and TDOS columns
energy = data[:, 0] 
tdos = data[:, 1]

# Create a plot
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed

# Plot the TDOS data
plt.plot(energy, tdos, label='TDOS', color='navy')

# Add a vertical dotted line for Fermi Energy
fermi_energy = 0.0  # Change this value to the actual Fermi Energy
plt.axvline(x=fermi_energy, color='gray', linestyle='--', label='Fermi Energy')

plt.xlabel('E - Ef (eV)', fontsize=16)
plt.ylabel('DOS (states/eV)', fontsize=16)
plt.title(f'{title}', fontsize=20)

# Add a dotted grid
plt.grid(True, linestyle='--')

plt.legend()

plt.show()  # Show the plot
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Load data from the file
data_O = np.loadtxt('dos_O.txt')  # Assumes a space-separated text file
data_C = np.loadtxt('dos_C.txt')  # Assumes a space-separated text file
title = "Density of States (DOS)"

# Extract Energy and TDOS columns
energy_O = data_O[:, 0] 
tdos_O = data_O[:, 1]

energy_C = data_C[:, 0] 
tdos_C = data_C[:, 1]

# Create a plot
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed

# Plot the TDOS data
plt.plot(energy_O, tdos_O, label='O', color='navy')
plt.plot(energy_C, tdos_C, label='C', color='maroon')

# Add a vertical dotted line for Fermi Energy
fermi_energy = 0.0  # Change this value to the actual Fermi Energy
plt.axvline(x=fermi_energy, color='gray', linestyle='--', label='Fermi Energy')

plt.xlabel('E - Ef (eV)', fontsize=16)
plt.ylabel('DOS (states/eV)', fontsize=16)
plt.title(f'{title}', fontsize=20)

# Add a dotted grid
plt.grid(True, linestyle='--')

plt.legend()

plt.show()  # Show the plot
```
